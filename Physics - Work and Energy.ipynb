# Work in Physics

In physics, work is defined as the process of a force causing a displacement of an object. It is only considered as work if the object is moved while the force is applied.

## Equation for Work

The most fundamental equation for work when the force is constant and parallel to the direction of displacement is:

$$ W = F \cdot d $$

where:

- $W$ is the work done on the object (in joules, J)
- $F$ is the magnitude of the constant force (in newtons, N)
- $d$ is the magnitude of the displacement (in meters, m)

## Work Along a Straight Line

When a force $\vec{F}$ causes a displacement $\Delta \vec{x}$ along a straight line, the work done $W$ can be expressed as:

$$ W = F_x \Delta x $$

where:

- $F_x$ is the x-component of the force
- $\Delta x = x_f - x_i$ is the displacement along the x-axis

The SI unit for work is the joule (J), which is equivalent to a newton-meter (N·m), or $kg \cdot m^2/s^2$.

## Work with Angle Consideration

If the force is applied at an angle $\theta$ to the displacement, the work done is given by:

$$ W = (F \cos \theta) d $$

where:

- $d$ is the magnitude of the displacement
- $\theta$ is the angle between the force vector $\vec{F}$ and displacement vector $\Delta \vec{x}$



## Positive and Negative Work

Work can be positive or negative:

- Positive if the force and displacement are in the same direction.
- Negative if they are in opposite directions.



## General Definition of Work

For a varying force, a more general definition of work is used, which involves integrating the force over the path of the displacement.


# Example 5.1: Work Done While Sledding Through the Yukon

## Scenario

A man pulls a sled loaded with salmon on ice, applying a force at an angle to the horizontal. The mass of the sled and salmon is 50.0 kg, and the man exerts a horizontal force of 120 N. The sled is pulled over a distance of 5.00 m, first with the force horizontal to the ground, and then at an angle of 30.0° to the horizontal. Additionally, there is a frictional force that eventually stops the sled.



## Objectives

1. Calculate the work done by the man on the sled for both angles (0° and 30°).
2. Determine the work done by the friction force.

## Work Calculation

### Part (a): When the force is horizontal $(θ = 0°)$

Using the work equation $ W = F \cdot d $ when the force is parallel to the displacement:

$$ W = F_x \Delta x = (120\, \text{N}) \cdot (5.00\, \text{m}) = 600\, \text{J} $$

### Part (b): When the force is at a 30° angle

The work done by a force at an angle $ \theta $ is $ W = (F \cos \theta) d $:

$$ W = (F \cos \theta) d = (120\, \text{N}) \cdot \cos(30°) \cdot (5.00\, \text{m}) = 520\, \text{J} $$

### Part (c): Work done by friction

The work done by friction is calculated using the force of friction and the displacement over which it acts:

$$ W_{\text{friction}} = F_{k} \Delta x = (45.0\, \text{N}) \cdot (18.2\, \text{m} - 12.4\, \text{m}) = -261\, \text{J} $$

*Note: The work done by friction is negative since the friction force acts in the opposite direction of displacement.*

## Remarks

- The normal force $ \vec{n} $, gravitational force $ m\vec{g} $, and the upward component of the applied force do no work on the sled as they are perpendicular to the displacement.
- The mass of the sled is not directly involved in the calculation of work done by the applied force but is relevant when considering frictional effects.


## Work Done by Friction and Net Work

Friction always dissipates energy, often in the form of heat or sound. When calculating the work done by friction, we consider its effect on mechanical energy alone.

### Example 5.2: More Sledding

#### Goal
Calculate the work done by friction and the net work when an object is acted on by an applied force.

#### Problem
Considering a sled with a mass of 50.0 kg and a coefficient of kinetic friction of 0.200 between the sled and snow:

(a) Calculate the work done by friction and the net work when the sled is pulled 5.00 m with a force of 120 N at 0° angle.
(b) Repeat the calculations for an applied force at an angle of 30.0°.

#### Strategy
Use the work done by friction and the work done by the applied force to determine the net work. The normal force is found using Newton’s second law.

#### Solution
(a) For a horizontal applied force:
- The normal force $ n = mg $.
- Work done by friction $ W_{\text{fric}} = -f_k \Delta x = -\mu_k n \Delta x = -\mu_k mg \Delta x = -(0.200)(50.0\, \text{kg})(9.80\, \text{m/s}^2)(5.00\, \text{m}) = -4900\, \text{J} $.
- Net work $ W_{\text{net}} = W_{\text{app}} + W_{\text{fric}} = 600\, \text{J} - 4900\, \text{J} = -4300\, \text{J} $.

(b) For an applied force at 30.0°:
- Normal force $ n = mg - F_{\text{app}} \sin \theta $.
- Work done by friction $ W_{\text{fric}} = -\mu_k (mg - F_{\text{app}} \sin \theta) \Delta x = -(0.200)(50.0\, \text{kg})(9.80\, \text{m/s}^2 - 120\, \text{N} \sin 30.0°)(5.00\, \text{m}) = -4300\, \text{J} $.
- Net work $ W_{\text{net}} = W_{\text{app}} + W_{\text{fric}} = 520\, \text{J} - 4300\, \text{J} = -3780\, \text{J} $.

#### Remarks
The applied force angle affects the work done on the sled. An optimal angle (11.3° in this case) results in the most net work for the same applied force.

## Kinetic Energy and the Work–Energy Theorem

Newton's second law can be complex, so an alternative approach is to relate the object's speed to the net work done on it.

### Kinetic Energy Definition

The kinetic energy (KE) of an object is defined as:

$$ KE = \frac{1}{2} mv^2 $$

where:
- $ m $ is the mass of the object.
- $ v $ is the speed of the object.

The unit of kinetic energy is the joule (J).

### Work–Energy Theorem

The net work done on an object is equal to the change in the object's kinetic energy:

$$ W_{\text{net}} = KE_{\text{f}} - KE_{\text{i}} = \Delta KE $$

This means that positive net work increases an object's speed, while negative net work decreases it.



### Application

For a car skidding to a stop:
- If the initial speed is doubled, the distance skidded increases fourfold.



## EXAMPLE 5.3: Collision Analysis

### Goal
To utilize the work-energy theorem to calculate stopping distance and impact speed in a vehicle braking scenario.

### Problem
A car with a mass of $1.00 \times 10^3 \text{kg}$ traveling at $35.0 \text{m/s}$ needs to brake to avoid a collision. The braking force due to kinetic friction is $8.00 \times 10^3 \text{N}$. Air resistance is ignored.

(a) Determine the minimum distance required to stop the car without collision.
(b) Calculate the speed of the car at collision if the initial distance is only $30.0 \text{m}$.

### Strategy
Use the work-energy theorem, where net work done by friction is equated to the change in kinetic energy. For part (a), assume final velocity $v_f = 0$ to find the stopping distance $D_x$. For part (b), solve for final velocity $v_f$ given the initial distance.

### Solution

#### Part (a): Minimum Stopping Distance
Apply the work-energy theorem:

$$ W_{\text{net}} = \frac{1}{2} m v_f^2 - \frac{1}{2} m v_i^2 $$

Since $v_f = 0$ and $W_{\text{net}} = -f_k D_x$, we have:

$$ -f_k D_x = -\frac{1}{2} m v_i^2 $$

Solving for $D_x$:

$$ D_x = \frac{v_i^2}{2 f_k/m} = \frac{(35.0 \text{m/s})^2}{2 \cdot (8.00 \times 10^3 \text{N} / (1.00 \times 10^3 \text{kg}))} = 76.6 \text{m} $$

#### Part (b): Speed at Impact
For a non-zero final velocity, rearrange the work-energy theorem to solve for $v_f$:

$$ v_f^2 = v_i^2 - \frac{2 f_k D_x}{m} $$

Plugging in the given values:

$$ v_f^2 = (35.0 \text{m/s})^2 - \frac{2 \cdot (8.00 \times 10^3 \text{N}) \cdot (30.0 \text{m})}{(1.00 \times 10^3 \text{kg})} = 745 \text{m}^2/\text{s}^2 $$

Therefore, the final velocity $v_f$ is:

$$ v_f = \sqrt{745 \text{m}^2/\text{s}^2} = 27.3 \text{m/s} $$

### Remarks
This example emphasizes the importance of maintaining a safe following distance and being prepared to stop, especially at high speeds where significant distance is covered even during the reaction time before braking begins.





## Conservative and Nonconservative Forces

### Conservative Forces
A conservative force, such as gravity, has the property that the work it does on an object is path-independent. That is, the work done by a conservative force when moving an object from one point to another is the same, regardless of the path taken between these two points.

### Nonconservative Forces
Nonconservative forces, like kinetic friction and air drag, dissipate energy. Work done against nonconservative forces cannot be easily recovered and is generally transformed into heat or sound.



### Example: Work Against Friction
The work required to slide a book from point $A$ to point $D$ is path-dependent if there is friction. Moving the book along a longer path requires more work due to the dissipative nature of friction.

### Work-Energy Theorem with Conservative and Nonconservative Forces
For a system where only conservative forces do work:

$$ W_{\text{nc}} + W_{\text{c}} = \Delta KE $$





Here, $W_{\text{nc}}$ is the work done by nonconservative forces, and $W_{\text{c}}$ is the work done by conservative forces.

## Gravitational Potential Energy

An object can have potential energy due to its position in a gravitational field. This potential energy can be converted into work.

### Gravitational Work and Potential Energy
The work done by gravity when an object falls from height $y_i$ to $y_f$ is:

$$ W_g = -mg(y_f - y_i) $$



This work is equivalent to the negative change in gravitational potential energy.

### Gravitational Potential Energy Definition
The potential energy ($PE$) at a height $y$ is given by:

$$ PE = mgy $$

### Work by Gravity and Potential Energy Change
The work done by gravity is equal to the negative change in potential energy:

$$ W_g = -(\Delta PE) = -[mg y_f - mg y_i] $$

### Extension of the Work-Energy Theorem
The work-energy theorem considering nonconservative forces and potential energy is:

$$ W_{\text{nc}} = \Delta KE + \Delta PE $$

### Reference Levels for Potential Energy
The reference level for potential energy can be chosen arbitrarily. The change in potential energy is what matters, not the absolute value.





## EXAMPLE 5.4: Wax Your Skis



### Goal
Calculate the change in gravitational potential energy for different choices of reference level.

### Problem
A 60.0 kg skier is at the top of a slope. At point $A$, she is 10.0 m vertically above point $B$.

### Strategy
The gravitational potential energy at the initial point $A$, PE_i, and at the final point $B$, PE_f, are calculated based on the chosen zero level for gravitational potential energy.

### Solutions

#### Part (a)
With the zero level at point $B$:

- The potential energy at point $A$ is:
$$ PE_i = mgy_i = (60.0\, \text{kg})(9.80\, \text{m/s}^2)(10.0\, \text{m}) = 5880\, \text{J} $$

- The potential energy at point $B$ is 0 J by the choice of the zero level.
- The change in potential energy as the skier goes from point $A$ to point $B$ is:
$$ \Delta PE = PE_f - PE_i = 0\, \text{J} - 5880\, \text{J} = -5880\, \text{J} $$

#### Part (b)
With the zero level at point $A$:

- The potential energy at point $A$ is 0 J by the choice of the zero level.
- The potential energy at point $B$ is:
$$ PE_f = mgy_f = (60.0\, \text{kg})(9.80\, \text{m/s}^2)(-10.0\, \text{m}) = -5880\, \text{J} $$
- The change in potential energy is the same:
$$ \Delta PE = PE_f - PE_i = -5880\, \text{J} - 0\, \text{J} = -5880\, \text{J} $$

#### Part (c)
With the zero level 2.00 m above point $B$:

- The potential energy at point $A$ is:
$$ PE_i = mgy_i = (60.0\, \text{kg})(9.80\, \text{m/s}^2)(8.00\, \text{m}) = 4704\, \text{J} $$
- The potential energy at point $B$ is:
$$ PE_f = mgy_f = (60.0\, \text{kg})(9.80\, \text{m/s}^2)(-2.00\, \text{m}) = -1176\, \text{J} $$
- The change in potential energy remains the same:
$$ \Delta PE = PE_f - PE_i = -1176\, \text{J} - 4704\, \text{J} = -5880\, \text{J} $$

### Remarks
The change in gravitational potential energy is $-5880\, \text{J}$, regardless of how the zero level is chosen.


## EXAMPLE 5.5: Platform Diver

### Goal
Use the conservation of energy to calculate the speed of a body falling in the presence of gravity.



### Problem
A diver with mass $m$ drops from a height of 10.0 m above the water's surface. Air resistance is negligible.

### Strategy
- The system includes the diver and the Earth.
- The conservation of mechanical energy principle applies since only gravity is doing work.
- Choose $y = 0$ at the water's surface.

### Solution

#### Part (a)
To find the diver's speed 5.00 m above the water’s surface:

1. The energy conservation equation is $KE_i + PE_i = KE_f + PE_f$.
2. Initially, $KE_i = 0$ and $PE_i = mgy_i$.
3. At 5.00 m above the water, the potential energy is $PE_f = mg \times 5.00\, \text{m}$.
4. Solve the energy conservation equation for $v_f$:

$$
0 + mgy_i = \frac{1}{2}mv_f^2 + mgy_f
$$

Canceling $m$ and solving for $v_f$ gives:

$$
v_f = \sqrt{2g(y_i - y_f)} = \sqrt{2 \times 9.80\, \text{m/s}^2 \times (10.0\, \text{m} - 5.00\, \text{m})} = 9.90\, \text{m/s}
$$

#### Part (b)
To find the diver’s speed as he hits the water:

1. Use the same conservation of energy principle with $y_f = 0$ m.
2. Solve for $v_f$:

$$
0 + mgy_i = \frac{1}{2}mv_f^2 + 0
$$

Again, canceling $m$ and solving for $v_f$ gives:

$$
v_f = \sqrt{2gy_i} = \sqrt{2 \times 9.80\, \text{m/s}^2 \times 10.0\, \text{m}} = 14.0\, \text{m/s}
$$

### Remarks
- The speed at halfway is not half the final speed due to the nature of gravitational acceleration.
- The final speed does not depend on the mass of the diver because the kinetic energy change of the Earth is negligible due to its large mass compared to the diver.


## EXAMPLE 5.6: The Jumping Bug

### Goal
Use conservation of mechanical energy and projectile motion concepts to calculate the launch speed of a grasshopper.

### Problem
A grasshopper jumps at an angle of 45° and reaches a maximum height of 1.00 m. Determine the initial speed $v_i$ with which it left the ground, neglecting air resistance.



### Strategy
- Apply conservation of energy between the initial launch point and the point at maximum height.
- Use the fact that the horizontal component of velocity remains constant in the absence of air resistance.

### Solution

1. At maximum height, the vertical component of velocity $v_y = 0$.
2. Conservation of energy equation:

$$ \frac{1}{2}mv_i^2 + mgy_i = \frac{1}{2}mv_x^2 + mgh $$

3. Since $y_i = 0$, the equation simplifies to:

$$ v_i^2 = v_x^2 + 2gh $$

4. The horizontal velocity component $v_x$ at maximum height is equal to the horizontal component of the initial velocity, which is $v_i \cos(45°)$ due to the launch angle.

5. Substitute $v_x = v_i \cos(45°)$ into the energy conservation equation and solve for $v_i$:

$$ v_i^2 = (v_i \cos(45°))^2 + 2gh $$

6. Simplify and solve for $v_i$:

$$ v_i^2 = \frac{1}{2}v_i^2 + 2g(1.00\, \text{m}) $$

7. Solve for $v_i$:

$$ v_i^2 = 2 \times 9.80\, \text{m/s}^2 \times 1.00\, \text{m} $$

$$ v_i = \sqrt{2 \times 9.80\, \text{m/s}^2 \times 1.00\, \text{m}} = 4.43\, \text{m/s} $$

### Remarks
The speed calculated is an indication of the grasshopper's relative strength. The answer demonstrates the significant speed grasshoppers can achieve during their jumps.

## EXAMPLE 5.8: Hit The Ski Slopes

### Goal
To apply the conservation of mechanical energy and work-energy theorem, taking into account friction on a horizontal surface.

### Problem
A skier starts from rest at a height of $20.0 \text{ m}$ on a frictionless slope and then encounters a flat surface with a kinetic friction coefficient of $0.210$. The task is to find (a) the skier's speed at the bottom of the incline, and (b) the distance traveled on the flat surface before coming to a stop.



### Strategy
- For part (a), use conservation of mechanical energy to calculate the speed at the bottom of the incline.
- For part (b), employ the work-energy theorem considering the work done by friction on the flat surface.

### Solution

#### Part (a): Speed at the Bottom
Calculate the skier's speed using the conservation of mechanical energy from the top to the bottom of the incline:

$$ v_B = \sqrt{2gh} = \sqrt{2 \cdot 9.80 \frac{\text{m}}{\text{s}^2} \cdot 20.0 \text{ m}} = 19.8 \frac{\text{m}}{\text{s}} $$

#### Part (b): Distance on the Horizontal Surface
The work done by friction is equal to the negative change in kinetic energy:

$$ W_{\text{net}} = W_{\text{fric}} = -f_k d = \Delta KE = \frac{1}{2}m v_C^2 - \frac{1}{2}m v_B^2 $$

Given that $ v_C = 0 $ (the skier comes to a stop), and $ f_k = \mu_k mg $, solve for the distance $d$:

$$ -\mu_k mgd = -\frac{1}{2}m v_B^2 $$

Solving for $d$ gives:

$$ d = \frac{v_B^2}{2\mu_k g} = \frac{(19.8 \frac{\text{m}}{\text{s}})^2}{2 \cdot 0.210 \cdot 9.80 \frac{\text{m}}{\text{s}^2}} = 95.2 \text{ m} $$

### Remarks
The distance $d$ that the skier travels on the horizontal surface is directly proportional to the height $h$ of the slope from which they descended, indicating that doubling the height of the slope would double the distance traveled on the flat surface.



## Spring Potential Energy and Conservation of Mechanical Energy

Springs are crucial components in various technologies, functioning in everything from small gadgets to vehicles. They are conservative forces, meaning the work done on or by a spring can be stored as potential energy.

### Hooke's Law

- A spring's force can be described by Hooke's law:

  $$ F_s = -kx $$

  where $k$ is the spring constant and $x$ is the displacement from the equilibrium position.

### Elastic Potential Energy

- The elastic potential energy ($PE_s$) stored in a spring is given by:

  $$ PE_s = \frac{1}{2}kx^2 $$

- This energy represents the work done to compress or stretch the spring from its equilibrium position.

### Work Done by a Spring

- The work done by a spring force when compressed from an initial displacement $x_i$ to a final displacement $x_f$ is:

  $$ W_s = \frac{1}{2}k(x_i^2 - x_f^2) $$

- This work can be incorporated into the work-energy theorem, leading to the equation:

  $$ W_{\text{nc}} + \frac{1}{2}kx_i^2 = \frac{1}{2}mv^2 + mgh + \frac{1}{2}kx_f^2 $$

  where $W_{\text{nc}}$ is the work done by nonconservative forces, $mv^2/2$ is the kinetic energy, and $mgh$ is the gravitational potential energy.

  

### Conservation of Mechanical Energy with Springs

- In the absence of nonconservative forces, the conservation of mechanical energy in systems involving springs and gravity can be expressed as:

  $$ (KE + PE_g + PE_s)_i = (KE + PE_g + PE_s)_f $$

  This indicates that the total mechanical energy (kinetic plus potential) remains constant over time.

### Application in Problem-Solving

- When addressing problems that include springs and gravitational forces, it is essential to define the spring's equilibrium point in addition to the zero level for gravitational potential energy.



## Spring Potential Energy and Conservation of Mechanical Energy

Springs are essential devices in various technologies, behaving according to conservative principles when compressed or stretched. The spring force follows Hooke's law, which states that the force ($F_s$) is proportional to the displacement ($x$):

$$ F_s = -kx $$

where $k$ is the spring constant.

### Elastic Potential Energy

The elastic potential energy ($PE_s$) in a spring when displaced from the equilibrium position is:

$$ PE_s = \frac{1}{2}kx^2 $$

This potential energy is considered stored energy resulting from the work done on the spring.

### Conservation of Mechanical Energy with Springs

In the absence of nonconservative forces, the mechanical energy of a system that includes a spring is conserved:

$$ (KE + PE_g + PE_s)_i = (KE + PE_g + PE_s)_f $$

where $KE$ is kinetic energy, $PE_g$ is gravitational potential energy, and $PE_s$ is the elastic potential energy.

### Example 5.9: A Horizontal Spring

#### Goal

Calculate the speed of a block on a horizontal spring, with and without friction.

#### Problem

A 5.00 kg block is attached to a spring (constant $k = 400$ N/m) on a frictionless surface. If the block is pulled out to $x_i = 0.050$ m and released, find the speed at different points.



#### Solution

(a) At the equilibrium point ($x_f = 0$), conservation of energy yields:

$$ \frac{1}{2}kx_i^2 = \frac{1}{2}mv_f^2 $$

Solving for $v_f$:

$$ v_f = \sqrt{\frac{k}{m}x_i^2} = \sqrt{\frac{400\, \text{N/m}}{5.00\, \text{kg}}(0.050\, \text{m})^2} = 0.447\, \text{m/s} $$

(b) At $x = 0.025$ m:

$$ \frac{1}{2}kx_i^2 = \frac{1}{2}mv_f^2 + \frac{1}{2}kx^2 $$

Solving for $v_f$:

$$ v_f = \sqrt{\frac{k}{m}(x_i^2 - x^2)} = \sqrt{\frac{400\, \text{N/m}}{5.00\, \text{kg}}((0.050\, \text{m})^2 - (0.025\, \text{m})^2)} = 0.387\, \text{m/s} $$

(c) With friction (coefficient $\mu_k = 0.150$), the work-energy theorem includes the work done by friction:

$$ -\mu_k mgx_i = \frac{1}{2}mv_f^2 - \frac{1}{2}kx_i^2 $$

The speed $v_f$ will be lower due to the energy dissipated by friction.

## EXAMPLE 5.10: Circus Acrobat

### Goal
To apply conservation of mechanical energy for a problem involving gravitational potential energy and spring potential energy.

### Problem
A 50.0-kg circus acrobat falls from a height of 2.00 m onto a springboard with a spring constant of $8000\, \text{N/m}$. The task is to calculate the maximum distance she compresses the spring.



### Strategy
- Use conservation of mechanical energy since there are no nonconservative forces acting.
- At the maximum compression of the spring, the acrobat's velocity is zero, thus kinetic energy is zero.
- Set $y = 0$ at the point of maximum compression, making the final gravitational potential energy zero.

### Solution

#### Part (a) Maximum Spring Compression
Using conservation of mechanical energy:

$$ (KE + PE_g + PE_s)_i = (KE + PE_g + PE_s)_f $$

The initial kinetic energy is zero, and the final gravitational potential energy is zero:

$$ 0 + mg(h + d) + 0 = 0 + 0 + \frac{1}{2}kd^2 $$

Rearrange and solve for the compression distance $d$:

$$ mg(h + d) = \frac{1}{2}kd^2 $$

Substitute the given values:

$$ (50.0\, \text{kg})(9.80\, \text{m/s}^2)(2.00\, \text{m} + d) = \frac{1}{2}(8000\, \text{N/m})d^2 $$

This results in a quadratic equation:

$$ 8000d^2 - (50.0 \times 9.80 \times 2.00)d - (50.0 \times 9.80)d^2 = 0 $$

Solving this equation yields:

$$ d = 0.560\, \text{m} $$

#### Remarks
- The negative solution for $d$ can be discarded because distance cannot be negative in this context.
- Changes in the acrobat’s center of mass, such as crouching, are not considered in this solution.






## EXAMPLE 5.11: A Block Projected Up A Frictionless Incline

### Goal
To utilize conservation of mechanical energy for solving problems on an incline involving gravitational potential energy and spring potential energy.

### Problem
A 0.500-kg block is compressed against a spring with a constant $k = 625\, \text{N/m}$ by $10.0\, \text{cm}$ and then released.
- (a) Determine the maximum distance $d$ the block travels up a $30.0^\circ$ incline.
- (b) Calculate the block's velocity at half its maximum height on the incline.



### Strategy
Apply conservation of mechanical energy for both parts of the problem since there are no nonconservative forces at play.

### Solution

#### Part (a)
To find the distance $d$ the block travels up the ramp:

1. Use the conservation of mechanical energy:

$$ \frac{1}{2}kx_i^2 = mgh = mgd \sin(\theta) $$

2. Solve for $d$:

$$ d = \frac{\frac{1}{2}kx_i^2}{mg \sin(\theta)} $$

3. Plug in the known values:

$$ d = \frac{\frac{1}{2}(625\, \text{N/m})(0.100\, \text{m})^2}{(0.500\, \text{kg})(9.80\, \text{m/s}^2) \sin(30.0^\circ)} = 1.28\, \text{m} $$

#### Part (b)
To find the velocity at half the maximum height $h/2$:

1. First, find the half height $h/2$:

$$ h = d \sin(\theta) = (1.28\, \text{m}) \sin(30.0^\circ) = 0.640\, \text{m} $$

2. Then, apply conservation of mechanical energy:

$$ \frac{1}{2}kx_i^2 = \frac{1}{2}mv_f^2 + mg\left(\frac{1}{2}h\right) $$

3. Solve for $v_f$:

$$ v_f = \sqrt{\frac{k}{m}x_i^2 - gh} = \sqrt{\frac{625\, \text{N/m}}{0.500\, \text{kg}}(0.100\, \text{m})^2 - (9.80\, \text{m/s}^2)(0.640\, \text{m})} = 2.50\, \text{m/s} $$

### Remarks
The problem illustrates that it is not necessary to calculate the velocity with which the block leaves the spring. Instead, the mechanical energy at the key points of interest suffices to determine the required quantities.


## 5.6 Systems and Energy Conservation

The work-energy theorem relates the work done by nonconservative ($W_{nc}$) and conservative forces ($W_c$) to the change in kinetic energy ($\Delta KE$) and potential energy ($\Delta PE$):

$$ W_{nc} + W_c = \Delta KE $$

The theorem can be expressed considering potential energy changes:

$$ W_{nc} = \Delta KE + \Delta PE = (KE_f - KE_i) + (PE_f - PE_i) $$

Rearranging this, we get:

$$ W_{nc} = (KE_f + PE_f) - (KE_i + PE_i) $$

If we define total mechanical energy ($E$) as $E = KE + PE$, then the work done by all nonconservative forces equals the change in mechanical energy of the system:

$$ W_{nc} = E_f - E_i = \Delta E $$

### Energy Transfer and Transformation

Energy may change form, transfer into or out of a system, but it cannot be created or destroyed. This is the principle of conservation of energy. Here are the ways energy can be transferred:

- **Work**: Displacement of a system by an applied force.
- **Heat**: Transfer of energy through microscopic collisions.
- **Mechanical Waves**: Energy transferred by disturbances through a medium.
- **Electrical Transmission**: Energy transferred through electric currents.
- **Electromagnetic Radiation**: Energy transferred in the form of waves like light or microwaves.

### Conservation of Energy in General

The conservation of energy principle is universal and not limited to physics. It applies to biological systems and cosmic events alike, showing that energy transformations are fundamental to all processes in the universe.

#### Example: The Impact of an Asteroid

An asteroid's kinetic energy could be immense due to its mass and high velocity. An impact would convert this kinetic energy into other forms, potentially causing massive devastation. This illustrates the power of kinetic energy and the significance of energy conservation on a cosmic scale.

### Implications

- **Biological Systems**: Chemical energy is transformed into mechanical energy or light within organisms.
- **Astronomical Events**: The kinetic energy of celestial bodies can have significant effects on Earth.
  
The conservation of energy is a key principle that explains the behavior of systems from microscopic to cosmic scales, highlighting the interconnectedness of different energy forms.



## 5.7 Power

Power is defined as the rate at which energy is transferred or work is done. It's a key concept in understanding the efficiency and performance of devices and living organisms in terms of energy transfer.

### Definition of Power

Power is the rate of doing work over a time interval $\Delta t$:

$$ P = \frac{W}{\Delta t} $$

where:
- $P$ is power,
- $W$ is work done,
- $\Delta t$ is the time interval.

The SI unit of power is the watt (W), defined as one joule per second (J/s):

$$ 1\, \text{W} = 1\, \text{J/s} = 1\, \text{kg} \cdot \text{m}^2/\text{s}^3 $$

### Power and Velocity

Power can also be expressed in terms of force and velocity:

$$ P = \frac{F \Delta x}{\Delta t} = Fv $$

where:
- $F$ is the force applied in the direction of the velocity $v$,
- $\Delta x$ is the displacement.

### Instantaneous Power

For varying forces and velocities, the instantaneous power is given by:

$$ P = Fv $$

In this equation, $F$ and $v$ must be parallel and can change with time.

### Units of Power

- **Horsepower (hp)**: Used to rate engine power, where $1\, \text{hp} = 550\, \text{ft-lb/s} = 746\, \text{W}$.
- **Kilowatt-hour (kWh)**: A unit of energy, not power, representing the energy transferred at a rate of 1 kW for 1 hour:

$$ 1\, \text{kWh} = (10^3\, \text{W})(3600\, \text{s}) = 3.6 \times 10^6\, \text{J} $$

### Application

The energy used by an electrical appliance can be calculated by multiplying its power rating by the time it operates. For example, a 100 W light bulb uses $0.100\, \text{kW} \times 1\, \text{h} = 0.1\, \text{kWh}$ or $3.6 \times 10^5\, \text{J}$ of energy in an hour.


## EXAMPLE 5.12: Power Delivered By An Elevator Motor

### Goal
To determine the power required by an elevator motor to lift a fully loaded elevator car against frictional forces at a constant speed using the force-times-velocity definition of power.

### Problem
An elevator car with a mass of $1000\, \text{kg}$ has a maximum load capacity of $800\, \text{kg}$. It experiences a constant frictional force of $4000\, \text{N}$ while moving upward. The motor must deliver enough power to lift the elevator at a constant speed of $3.00\, \text{m/s}$.



### Strategy
The power delivered by the motor is calculated by determining the tension in the elevator cable and using the relationship $P = Fv$, where $v$ is the constant velocity of the elevator.

### Solution

1. Use Newton's second law for the elevator car with acceleration $a = 0$ due to constant velocity:

$$ \sum F = m \cdot a $$

2. The forces on the elevator car are tension ($T$), friction ($f$), and gravity ($Mg$), where $M$ is the total mass of the elevator and its load.

3. Balance the forces (tension up, friction and gravity down):

$$ T - f - Mg = 0 $$

4. Solve for the tension $T$:

$$ T = f + Mg $$

5. Plug in the values to get $T$:

$$ T = 4000\, \text{N} + (1000\, \text{kg} + 800\, \text{kg}) \cdot 9.8\, \text{m/s}^2 $$

6. Calculate the power $P$ using the force-velocity relationship:

$$ P = T \cdot v $$

7. Convert the power to kilowatts and horsepower:

$$ P(\text{W}) = P(\text{J/s}) \cdot \frac{1}{1000} \quad (\text{to convert to kW}) $$
$$ P(\text{hp}) = P(\text{W}) \cdot \frac{1}{746} \quad (\text{to convert to hp}) $$

### Remarks
Friction increases the power required by the motor. Conversely, for a descending elevator, friction could reduce the power needed.


## EXAMPLE 5.13: Shamu Sprint

### Goal
To calculate the average power needed for a killer whale to significantly increase its kinetic energy.

### Problem
A killer whale named Shamu, with a mass of $8000\, \text{kg}$, accelerates to $12.0\, \text{m/s}$ in $6.00\, \text{s}$, ignoring water drag. Determine the average power required.

### Strategy
Use the change in kinetic energy to determine the work done, then divide by the time interval to find the average power.

### Solution

1. Calculate the change in kinetic energy:

$$ \Delta KE = \frac{1}{2}mv_f^2 - \frac{1}{2}mv_i^2 $$

2. Since Shamu starts from rest ($v_i = 0$), the change in kinetic energy is:

$$ \Delta KE = \frac{1}{2} \cdot 8000\, \text{kg} \cdot (12.0\, \text{m/s})^2 = 5.76 \times 10^5\, \text{J} $$

3. Use the definition of power where work done $W = \Delta KE$:

$$ P = \frac{W}{\Delta t} = \frac{5.76 \times 10^5\, \text{J}}{6.00\, \text{s}} = 9.60 \times 10^4\, \text{W} $$

### Remarks

- This calculation provides the minimum power needed, not accounting for additional resistance due to water and internal muscle friction.
- The computed power is substantial, enough to supply a moderate-sized office building.
- The principle can be similarly applied to gravitational potential energy changes.



## EXAMPLE 5.14: Speedboat Power

### Goal
To determine the power requirements for a speedboat to accelerate in the presence of drag force and to express instantaneous power in terms of drag force, mass, acceleration, and time.

### Problem
(a) Calculate the average power required for a $1000\, \text{kg}$ speedboat to accelerate from rest to $20.0\, \text{m/s}$ in $5.00\, \text{s}$ against a constant drag force of $500\, \text{N}$. 
(b) Derive an expression for instantaneous power incorporating drag force $f_d$, mass $m$, acceleration $a$, and time $t$.

### Strategy
Incorporate work done by the engine and drag force into the work-energy theorem, and use kinematics for acceleration and displacement.

### Solution

#### Part (a)
1. Apply the work-energy theorem, with initial velocity $v_i = 0$:

$$ W_{\text{net}} = \Delta KE = \frac{1}{2} m v_f^2 $$

2. Calculate the acceleration $a$:

$$ v_f = a t \Rightarrow a = \frac{v_f}{t} = \frac{20.0\, \text{m/s}}{5.00\, \text{s}} = 4.00\, \text{m/s}^2 $$

3. Find the displacement $\Delta x$:

$$ v_f^2 = 2a \Delta x \Rightarrow \Delta x = \frac{v_f^2}{2a} = \frac{(20.0\, \text{m/s})^2}{2 \cdot 4.00\, \text{m/s}^2} = 50.0\, \text{m} $$

4. Determine work done against drag force:

$$ W_{\text{drag}} = -f_d \Delta x = -(500\, \text{N})(50.0\, \text{m}) = -2.50 \times 10^4\, \text{J} $$

5. Solve for work done by the engine:

$$ W_{\text{engine}} = \frac{1}{2} m v_f^2 - W_{\text{drag}} = \frac{1}{2} (1000\, \text{kg}) (20.0\, \text{m/s})^2 - (-2.50 \times 10^4\, \text{J}) $$

6. Calculate average power:

$$ P = \frac{W_{\text{engine}}}{\Delta t} = \frac{2.25 \times 10^5\, \text{J}}{5.00\, \text{s}} = 4.50 \times 10^4\, \text{W} \quad \text{or} \quad 60.3\, \text{hp} $$

#### Part (b)
Derive the expression for instantaneous power:

1. Apply Newton's second law:

$$ m a = F_E - f_d $$

2. Solve for the engine force $F_E$:

$$ F_E = m a + f_d $$

3. Substitute $F_E$ and $v = at$ into the power equation:

$$ P = F_E v = (m a + f_d)(a t) = m a^2 t + f_d a t $$

### Remarks
Actual drag forces tend to increase with speed, which would affect the power requirement over time.


## 5.7.1 Energy and Power in a Vertical Jump

A vertical jump consists of two phases: extension and free flight. The motion is characterized by the center of mass (CM), which simplifies the complex motion of different body parts into a single point.

### Conservation of Mechanical Energy in Jumping

To find the maximum height $H$ that the CM rises, we apply the principle of conservation of mechanical energy. At liftoff, the gravitational potential energy $PE_i$ is taken to be zero, and at the peak, the kinetic energy $KE_f$ is zero.

The conservation equation is:

$$ PE_i + KE_i = PE_f + KE_f $$

Which simplifies to:

$$ \frac{1}{2}mv_{CM}^2 = mgH $$

Solving for $H$ gives us:

$$ H = \frac{v_{CM}^2}{2g} $$

### Estimating the Liftoff Velocity $v_{CM}$

Assuming constant acceleration $a$ during the extension phase, and given the depth of crouch $h$ and time $\Delta t$ for extension, we can estimate $v_{CM}$ as:

$$ v_{CM} = \frac{2h}{\Delta t} $$

With typical values $h = 0.40\, \text{m}$ and $\Delta t = 0.25\, \text{s}$, we get:

$$ v_{CM} = 3.2\, \text{m/s} $$

Substituting into the equation for $H$, we find:

$$ H = \frac{(3.2\, \text{m/s})^2}{2 \cdot 9.80\, \text{m/s}^2} = 0.52\, \text{m} $$

### Energy and Power Considerations

The kinetic energy imparted to the body during a jump is:

$$ KE = \frac{1}{2} m v_{CM}^2 $$

For a person with mass $68\, \text{kg}$:

$$ KE = \frac{1}{2} (68\, \text{kg}) (3.2\, \text{m/s})^2 = 3.5 \times 10^2\, \text{J} $$

Considering muscle efficiency of 25%, the chemical energy used is approximately $4 \times 350\, \text{J} = 1400\, \text{J}$, which equates to about one-third of a food calorie.



### Average Mechanical Power Output

The average mechanical power generated during the jump can be calculated as:

$$ P = \frac{KE}{\Delta t} = \frac{3.5 \times 10^2\, \text{J}}{0.25\, \text{s}} = 1.4 \times 10^3\, \text{W} $$

Converting to horsepower:

$$ P \approx 1.9\, \text{hp} $$
